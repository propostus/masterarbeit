```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Pfad zur Datei, die du debuggen willst
file_path = "../audio_files/common_voice/raw/cv-corpus-de-combined-20-21-delta/common_voice_de_41249729.mp3"

# Audio laden
audio, sr = librosa.load(file_path, sr=16000)

# Plotten des Wellenformsignals
plt.figure(figsize=(14, 4))
librosa.display.waveshow(audio, sr=sr)
plt.title("Waveform")
plt.xlabel("Zeit (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()

# Zustzliche Debug-Ausgabe
print(f"Min: {np.min(audio)} | Max: {np.max(audio)} | Mean: {np.mean(audio)}")
print(f"Shape: {audio.shape}, Dauer: {len(audio)/sr:.2f}s")
```
