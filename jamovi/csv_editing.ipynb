# V2: Spectral Contrast in einzelne Spalten aufteilen

```python
import pandas as pd

# Lade CSV
df = pd.read_csv("features_and_wer_10h.csv")

# Speichere alte Spaltenreihenfolge
cols = df.columns.tolist()

# Finde Index von "spectral_contrast"
insert_at = cols.index("spectral_contrast")

# Entferne Klammern und splitte
df["spectral_contrast"] = df["spectral_contrast"].astype(str).str.strip("[]")
contrast_split = df["spectral_contrast"].str.split(expand=True).astype(float)
contrast_split.columns = [f"spectral_contrast_{i+1}" for i in range(contrast_split.shape[1])]

# Entferne alte Spalte und füge neue an ursprünglicher Stelle ein
df = df.drop(columns=["spectral_contrast"])
df = pd.concat([df.iloc[:, :insert_at], contrast_split, df.iloc[:, insert_at:]], axis=1)

# Speichern
df.to_csv("features_and_wer_10h_v2.csv", index=False)
```

# V3: MFCC-Spalten umbenennen

```python
import pandas as pd

# Lade den DataFrame
df = pd.read_csv("features_and_wer_10h_v2.csv")

# Parameter
num_mfcc = 13
stats = ["mean", "std", "min", "max", "range", "skewness", "kurtosis"]

# Neue sprechende Spaltennamen erzeugen
new_column_names = []
for stat in stats:
    for i in range(1, num_mfcc + 1):  # 1-indiziert
        new_column_names.append(f"mfcc_{i}_{stat}")

# Alte Spaltennamen (basierend auf 1-indizierter Konvention)
old_column_names = [f"mfcc_stat_{i}" for i in range(1, len(new_column_names) + 1)]

# Ersetze alte durch neue Namen
rename_map = {old: new for old, new in zip(old_column_names, new_column_names)}
df.rename(columns=rename_map, inplace=True)

# Optional: Speichern
df.to_csv("features_and_wer_10h_v3.csv", index=False)
```
