# WER-Analyse – Common Voice 10h

- Datum: 2025-06-14
- Modelle: Whisper tiny, base, small, medium, large
- Datensatz: common_voice_subset_10h


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Layout
sns.set(style="whitegrid")

# CSV laden
df = pd.read_csv("../../results/subset_10h/features_and_wer_10h.csv")
df.shape
```

```python
# WER-Spalten identifizieren
wer_cols = [col for col in df.columns if col.startswith("wer_")]

# Feature-Spalten bereinigen (ohne chroma & mfcc)
excluded_prefixes = ("mfcc_stat_", "chroma_")
feature_cols = [col for col in df.select_dtypes(include="number").columns 
                if not col.startswith(excluded_prefixes) and not col.startswith("wer_")]

# Korrelation nur mit ausgewählten Features
corr_with_wer = df[feature_cols + wer_cols].corr().loc[feature_cols, wer_cols]
corr_with_wer = corr_with_wer.sort_values(by="wer_tiny")

# Anzeigen
corr_with_wer.head(24)
```
