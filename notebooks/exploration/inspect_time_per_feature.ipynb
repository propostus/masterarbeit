```python
import pandas as pd

df = pd.read_csv("../results/feature_times.csv")
df.head()
```

```python
print(f"Anzahl Dateien: {df.shape[0]}")
print(f"Anzahl Spalten: {df.shape[1]}")
df.columns.tolist()
```

```python
# Alle Spalten, die mit "time_" beginnen
time_cols = [col for col in df.columns if col.startswith("time_")]
df_times = df[time_cols]
df_times.head()
```

```python
# Durchschnittliche Berechnungszeit pro Feature (in Sekunden)
mean_times = df_times.mean().sort_values(ascending=False)
mean_times
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
mean_times.plot(kind="bar")
plt.ylabel("Durchschnittliche Dauer (s)")
plt.title("Durchschnittliche Berechnungszeit je Feature")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
df["total_time"] = df[time_cols].sum(axis=1)
df[["filename", "duration_sec", "total_time"]].head(10)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df["total_time"], bins=10, color="steelblue", edgecolor="black")
plt.xlabel("Gesamtdauer [s]")
plt.ylabel("Anzahl Dateien")
plt.title("⏱ Gesamt-Berechnungszeit pro Datei")
plt.grid(True)
plt.show()
```

```python
df["relative_compute_cost"] = df["total_time"] / df["duration_sec"]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(df["relative_compute_cost"], bins=20, color="orange", edgecolor="black")
plt.xlabel("Verhältnis: total_time / duration_sec")
plt.ylabel("Anzahl Dateien")
plt.title("⏱ Verhältnis Berechnungszeit zu Audiodauer")
plt.grid(True)
plt.tight_layout()
plt.show()
```
