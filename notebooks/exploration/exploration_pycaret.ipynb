```python
from pycaret.regression import *
```

```python
import pandas as pd

# Pfad zu deinem kombinierten Datensatz
df = pd.read_csv("../../results/subset_10h/features_and_wer_10h.csv")

# Optional: Überblick prüfen
df.head()
```

```python
reg_setup = setup(
    data=df,
    target='wer_tiny',
    session_id=42,
    normalize=True,
    ignore_features=[
        "filename", "filepath", "reference",
        "hyp_tiny", "hyp_base", "hyp_small", "hyp_medium", "hyp_large",
        "wer_base", "wer_small", "wer_medium", "wer_large"
    ] + 
    [col for col in df.columns if col.startswith("chroma_") and col != "chroma_1"] + 
    [col for col in df.columns if col.startswith("mfcc_stat_") and col != "mfcc_stat_1"],
    verbose=False
)
```
