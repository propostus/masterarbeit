```python
import pandas as pd
import os

# Anpassen: ein Verzeichnis hoch
base_dir = "../results/de-20-21-delta"
wer_dir = os.path.join(base_dir, "wer")

# WER-Dateien laden
wer_20 = pd.read_csv(os.path.join(wer_dir, "cv-de-20_wer.csv"))
wer_21 = pd.read_csv(os.path.join(wer_dir, "cv-de-21_wer.csv"))

# ZusammenfÃ¼hren
wer_combined = pd.concat([wer_20, wer_21], ignore_index=True)

# Features laden
features_path = os.path.join(base_dir, "features-20-21-delta-250907.csv")
features_df = pd.read_csv(features_path)

# Merge (via filename)
merged_df = pd.merge(features_df, wer_combined, on="filename", how="inner")

# Speichern
output_path = os.path.join(base_dir, "features_wer_de_20_21_combined.csv")
merged_df.to_csv(output_path, index=False)

print(f"âœ… Merge erfolgreich! Gespeichert unter:\n{output_path}")
print(f"ðŸ”¢ Anzahl kombinierter Samples: {len(merged_df)}")
```
